## Problem 3.

Run the following code which explains how to make a kind of animation effect in Matlab.

```octave
%1)

figure(1); hold on
axis([-1 1 -4 .5])

for x = -1:0.05:1
    y = -4*x^2;
    plot(x,y,'s')
    pause(0.1)
end
hold off
```

## Problem 4.

Write a code to create a circle as shown in the below figure. (hint: there are many ways to realize this problem, but think about Eulerâ€™s formula (http://en.wikipedia.org/wiki/Euler's_formula))  

```octave
y = -1:0.01:1;
x1 = sqrt(1 - y.^2);
x2 = -sqrt(1 - y.^2);

plot(y,x1), hold on
plot(y,x2), hold off
axis equal
```

```octave
center = [0 0];                 % ì¤‘ì‹¬ì 
r = 70;                         % ë°˜ì§€ë¦„
N = 1000;                       % ì›ì˜ ì  ê°œìˆ˜
NL = 600;                       % ì„ ì˜ ì  ê°œìˆ˜*2
theta = linspace(0,2*pi,N);     % ì›(ë¼ë””ì•ˆ)
x = r*cos(theta) + center(1);       
y = r*sin(theta) + center(2);       
plot(x,y);                      

axis equal;                     % ì¢Œí‘œì¶•.
grid on
```

```octave
t = [0:0.1:2*pi];

% í‚¤ë³´ë“œë¡œë¶€í„° ìž…ë ¥í•œë‹¤

r  = 1;
x0 = 0;
y0 = 0;

% ì‹œê°„ t=0~2pi ê¹Œì§€ ëŒë¦¬ë©´ì„œ ì›ì˜ xì¢Œí‘œì™€ yì¢Œí‘œë¥¼ ë½‘ì•„ë‚¸ë‹¤.=> ë°˜ë³µë¬¸ ì‚¬ìš©: forë¬¸

for j = 1:length(t),
    x(j) = x0 + r*cos(t(j));
    y(j) = y0 + r*sin(t(j));
end

% ê·¸ëž˜í”„ë¥¼ ê·¸ë¦°ë‹¤

plot(x,y)
title('Circle')
xlabel('x(t)')
ylabel('y(t)')
```

## Problem 5.

Create an animation which the green dot is moving around the unit circle (counter clockwise). 

```octave
[x,y] = scircle1(0,0,1);

for y1 = -1:0.01:1;
    x1 = sqrt(1 - y1.^2);
    plot(x1,y1,'s',x,y)
    pause(0.01)
end

for y2 = -1:0.01:1;
    x2 = -sqrt(1 - y2.^2);
    plot(x2,-y2,'s',x,y)
    pause(0.01)
end
```
